```python
# OOPS
'''
a. Encapsulation
    - make private variables .So that , they cannot be modified from outside until they are accessed first.
b. Abstraction
    - hide the process of actions which are not intended to expose.
c. Inheritence
    - inherit the properties and methods of its parents.
d. Polymorphism
    - many form, same method or property can have many form for each instance.
'''
```

```python
from abc import ABC, abstractmethod
class Human(ABC):
    hobby = 'walking' # class property/attribute/variable
    def __init__(self, name, age): # Constructor
        self.name = name # instance property/attribute/variable
        self.age = age
    
    def talk(self, word): # instance method
        return f'Hello I am {self.name}, {word}'
        # return f'Hello {word}'
    
    @classmethod # classmethod is used to access class property
    def get_hobby(cls):
        return cls.hobby
    
    @staticmethod # staticmethod is used when some processing is required related to class but class or instance are not required at all.
    def get_planet_name(name):
        pass

    @abstractmethod # child class must define the abstract method during inheritance
    def disp():
        pass
    
    # Accessor method => use to access the property for reading purpose only.
    # Mutator method => use to change the value of property.

class Male(Human):
    def __init__(self, name, age, year):
        super().__init__(name, age)
        self.year = year
    
    def disp():
        return 'I am disp'
    
# himanshu = Human('Himanshu', 23)
# Human.hobby
# Human.get_hobby()
# himanshu.hobby

saket = Male('Saket', 25, 2026)
saket.talk('hello')

```
